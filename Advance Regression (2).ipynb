```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error

# Assuming you have a DataFrame 'df' and target variable 'y'

# Split data (if not already done)
X_train, X_test, y_train, y_test = train_test_split(df.drop('SalePrice', axis=1), df['SalePrice'], test_size=0.2, random_state=42)

# Define a grid of alpha values to explore
alpha_grid = np.logspace(-5, 0, 6)  # Explore a range of alpha values (adjust as needed)

# Define and fit the GridSearchCV model
model = Lasso()
grid_search = GridSearchCV(estimator=model, param_grid={'alpha': alpha_grid}, cv=5, scoring='neg_mean_squared_error')  # Minimize mean squared error
grid_search.fit(X_train, y_train)

# Access the best model and best alpha
best_model = grid_search.best_estimator_
best_alpha = grid_search.best_params_['alpha']

# Evaluate the best model on testing set
y_pred = best_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R-squared (best model): {r2:.4f}")
print(f"Mean Squared Error (best model): {mse:.4f}")

# You can now use the 'best_model' for further analysis or prediction on new data
```

```python
import matplotlib.pyplot as plt

def plot_correlation_matrix(data, title="Correlation Matrix"):
  """
  This function creates a correlation matrix heatmap visualization.

  Args:
      data (pandas.DataFrame): The data containing the variables.
      title (str, optional): The title for the plot. Defaults to "Correlation Matrix".
  """
  # Calculate correlation coefficients
  correlation_matrix = data.corr()

  # Create a new figure for plotting
  plt.figure(figsize=(10, 8))

  # Create heatmap with colorbar
  ax = plt.matshow(correlation_matrix, cmap="coolwarm")
  plt.colorbar(ax, label="Correlation Coefficient")

  # Set ticks and labels for the axes
  plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
  plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

  # Add text labels for each cell (optional)
  for (i, j), z in np.ndenumerate(correlation_matrix):
    plt.text(j, i, "{:0.2f}".format(z), ha="center", va="center", fontsize=8)

  # Set plot title and show the plot
  plt.title(title)
  plt.show()

# Example usage (assuming you have your data in a pandas DataFrame 'df')
plot_correlation_matrix(df)
```
